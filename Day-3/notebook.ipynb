```python
import pandas as pd
```

```python
from google.colab import files
import os

# Check if the file exists, if not, prompt for upload
file_name = 'Titanic-Dataset.csv'
if not os.path.exists(file_name):
  uploaded = files.upload()
  if file_name not in uploaded:
    print(f"Warning: '{file_name}' not uploaded. Please upload the file.")
  else:
    print(f"'{file_name}' uploaded successfully.")

try:
  titanic_data = pd.read_csv(file_name)
  print("Dataset loaded successfully.")
  print(titanic_data.head())
  print('\n')

  print(titanic_data.info())
  print('\n')

  print(titanic_data.describe())

except FileNotFoundError:
  print(f"Error: The file '{file_name}' was not found. Please ensure it is uploaded.")
except Exception as e:
  print(f"An error occurred: {e}")
```

```python
# Data Cleaning

titanic_data['Age'] = titanic_data['Age'].fillna(titanic_data['Age'].median())

embarked_mode = titanic_data["Embarked"].mode()
if not embarked_mode.empty:
    titanic_data["Embarked"] = titanic_data["Embarked"].fillna(embarked_mode[0])
else:
    titanic_data["Embarked"] = titanic_data["Embarked"].fillna('S')

if 'Cabin' in titanic_data.columns:
    titanic_data.drop(columns=['Cabin'], inplace=True)

print(titanic_data.isnull().sum())
```

# Data Analysis

```python
survival_gender = titanic_data.groupby('Sex')['Survived'].mean()
survival_class = titanic_data.groupby('Pclass')['Survived'].mean()
avg_age_per_class = titanic_data.groupby('Pclass')['Age'].mean()

print(survival_gender, '\n')
print(survival_class, '\n')
print(avg_age_per_class, '\n')

# Age Groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child', 'Teenager', 'Adult', 'Middle-aged', 'Senior']
titanic_data['AgeGroup'] = pd.cut(titanic_data['Age'], bins=bins, labels=labels)

age_group_survival = titanic_data.groupby('AgeGroup', observed=True)['Survived'].mean()
print(age_group_survival)
```

# Filtering

```python
# female servivors
female_servivors = titanic_data[(titanic_data['Sex']=='female') & (titanic_data['Survived']==1)]
print(female_servivors)
```

```python
# chlidren servivors
children_servivors = titanic_data[(titanic_data['Age'] < 18 ) & (titanic_data['Survived']==1)]
print(children_servivors)
```

```python

# first class servivors
first_class_servivors = titanic_data[(titanic_data['Pclass'] == 1) & (titanic_data['Survived']==1)]
print(first_class_servivors)
```

# Insights


1. Who was more likely to survive? (Gender)

Females were significantly more likely to survive.

* Female survival rate: ≈74.2%

* Male survival rate: ≈18.9%

The "women and children first" protocol is clearly visible in these numbers.


2. Did class affect survival?

Yes, there was a direct correlation between social class and survival probability. Passengers in higher classes had much better survival rates:

* 1st Class: ≈63.0%

* 2nd Class: ≈47.3%

* 3rd Class: ≈24.2%

# 4. What combination had the highest survival rate?

The combination with the absolute highest survival rate (100%) was:

* Female children in 2nd Class * Male children in 1st Class

Following closely behind were Adult females in 1st Class, who had a survival rate of approximately 97.4%
